```python
from google.colab import files
from PyPDF2 import PdfReader
import pytesseract
from PIL import Image
from pdf2image import convert_from_bytes
import io

uploaded_files = files.upload()

# Iterate through uploaded files
for pdf_filename, pdf_content in uploaded_files.items():
    # Convert bytes-like object to file-like object
    pdf_content_io = io.BytesIO(pdf_content)

    # Use PyPDF2 to read the number of pages in the PDF
    pdf_reader = PdfReader(pdf_content_io)
    num_pages = len(pdf_reader.pages)

    # Extract text from each page using Tesseract OCR
    for page_num in range(num_pages):
        # Convert the PDF page to an image
        images = convert_from_bytes(pdf_content, 500, first_page=page_num + 1, last_page=page_num + 1)

        # Extract text from the image using Tesseract OCR
        extracted_text = pytesseract.image_to_string(images[0], lang='eng')

        # Print the extracted text
        #print(f"Text from Page {page_num + 1} of {pdf_filename}:\n{extracted_text}\n")
```
